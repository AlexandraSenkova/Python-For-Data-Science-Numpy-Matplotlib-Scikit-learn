## Задание 2

С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

```python
t = np.linspace(0, 10, 51)
```

Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.

```python
f = np.cos(t)
# добавим на график вспомогательные прямые, ограничивающие область значений ф-ии f = cos(t)
f1 = 1 + t * 0
f2 = -1 + t * 0
# # для функций f1 и f2 добавлено слагаемое "t * 0", чтобы прямые построились на нашем графике.
```

Постройте линейную диаграмму, используя 
массив t для координат по горизонтали,а 
массив f - для координат по вертикали. 
Линия графика должна быть зеленого цвета.

Выведите название диаграммы - 'График f(t)'. 
Также добавьте названия 
для горизонтальной оси - 'Значения t' и 
для вертикальной - 'Значения f'.
Ограничьте график 
по оси x значениями 0.5 и 9.5, 
а по оси y - значениями -2.5 и 2.5.

```python
plt.plot(t, f)
plt.plot(t, f1, color="red", linestyle="dashed")
plt.plot(t, f2, color="red", linestyle="dashed")

# plt.plot(t, f, 'g-.o', label="t:[0;10]", color="blue", linestyle="dashed")
# plt.plot(t, f, 'b-s', label="t:[0;10]", color="blue", linestyle="dotted")
# o - circle
# s - square
# v, <, >, ^ - triangles pointing down, left, right, up
# d, D - thin or thick diamond
# x, X - cross (x is line only, X is filled shape)
# h - hexagon

# зададим заголовок и подписи, для формата и шрифтов используем аргумент fontdict
plt.title("График f = cos(t)", fontdict=title_font, loc="center")
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)

# зададим параметры границ для оси x и по оси y с помощью функции plt.axis: 
plt.axis([0.5, 9.5, -2.5, 2.5])

# дополнительно зададим легенду
legend = plt.legend(labels=["f = cos(t)", "f = 1", "f = -1"],loc = 'lower right', prop=legend_font, frameon = True)
plt.setp(legend.get_texts(), color="DarkBlue")

# нанесем сетку
plt.grid()
# plt.grid(axis="y", color="lightgrey")

plt.show()
```

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)

```python
x = np.linspace(-3, 3, 51)

y1 = x**2 
y2 = 2 * x + 0.5 
y3 = -3 * x - 1.5 
y4 = np.sin(x)
```

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, 
чтобы у вас было 4 отдельных графика в сетке, состоящей из 
двух строк и двух столбцов. 
В каждом графике массив x используется для координат по горизонтали.
В левом верхнем графике для координат по вертикали используйте y1, 
в правом верхнем - y2, 
в левом нижнем - y3, 
в правом нижнем - y4.
Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax[0,0].plot(x, y1)
ax[0,0].set_title("График y1")
ax[0,0].set_ylabel("y1", fontdict=label_font)
ax[0,0].set_xlim([-5, 5])
ax[0,1].plot(x, y2)
ax[0,1].set_title("График y2")
ax[0,1].set_ylabel("y2", fontdict=label_font)
ax[1,0].plot(x, y3)
ax[1,0].set_title("График y3")
ax[1,0].set_ylabel("y3", fontdict=label_font)
ax[1,1].plot(x, y4)
ax[1,1].set_title("График y4")
ax[1,1].set_ylabel("y4", fontdict=label_font)

```

```python
# другой способ:
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# расположим элементы двумерного массива последовательно друг за другом 
# с помощью метода .flatten и получим одномерный, что позволит 
# обращаться к элементам массива 'ax' без индексов:

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_ylabel("y1", fontdict=label_font)
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title("График y2")
ax2.set_ylabel("y2", fontdict=label_font)
ax3.plot(x, y3)
ax3.set_title("График y3")
ax3.set_ylabel("y3", fontdict=label_font)
ax4.plot(x, y4)
ax4.set_title("График y4")
ax4.set_ylabel("y4", fontdict=label_font)
```
